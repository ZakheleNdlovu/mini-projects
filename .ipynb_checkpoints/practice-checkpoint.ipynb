```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = pd.read_csv('gas_prices.csv')
b = a['Year']
c = a['Italy']
```

```python
plt.plot(b,c)
plt.title('Gas Prices in Italy')
plt.xlabel('Years')
plt.ylabel('Gas Prices')
```

```python
d = a['USA']
plt.plot(b,d)
plt.title('USA gas prices')
plt.xlabel('Years')
plt.ylabel('Gas Prices')
```

```python
e = a['Germany']
plt.plot(b,e)
plt.title('Germany gas prices')
plt.xlabel('Years')
plt.ylabel('Gas Prices')
```
